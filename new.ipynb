```python
import matplotlib.pyplot as plt
```

```python
x=["python","java", "c++"]
y=[85, 70 ,60]
plt.bar(x,y)
plt.show()
```

```python
x=["python","java", "c++"]
y=[85, 70 ,60]
c=["y","b","g"]
plt.bar(x,y,width=0.4,color=c,align= "edge")

plt.xlabel("programming language",fontsize=11)
plt.ylabel("No.")
plt.title("Performance")
plt.show()
```

```python
x=["python","java", "c++"]
y=[85, 70 ,60]
Z=[20,30,40]
plt.bar(x,y,width=0.4,color="r",label="popularity")
plt.bar(x,Z,width=0.4,color="b",label="popularity1")
plt.xlabel("programming language",fontsize=11)
plt.ylabel("No.")
plt.title("Performance")
plt.show()
```

```python
x=["python","java", "c++"]
y=[85, 70 ,60]
Z=[20,30,40]
p=np.arange(len(x))
plt.bar(p,y,width=0.4,color="r",label="popularity")
plt.bar(p,Z,width=0.4,color="b",label="popularity1")
plt.xlabel("programming language",fontsize=11)
plt.ylabel("No.")
plt.title("Performance")
plt.show()
```

```python
x=["python","java", "c++"]
y=[85, 70 ,60]
Z=[20,30,40]

p=np.arange(len(x))
p1=[i+0.4 for i in p]

plt.bar(p,y,width=0.4,color="r",label="popularity")
plt.bar(p1,Z,width=0.4,color="b",label="popularity1")

plt.xlabel("programming language",fontsize=11)
plt.ylabel("No.")
plt.title("Performance")
plt.show()
```

```python
x=["python","java", "c++"]
y=[85, 70 ,60]
Z=[20,30,40]

p=np.arange(len(x))
p1=[i+0.4 for i in p]

plt.bar(p,y,width=0.4,color="r",label="popularity")
plt.bar(p1,Z,width=0.4,color="b",label="popularity1")

plt.xlabel("programming language",fontsize=11)
plt.ylabel("No.")
plt.title("Performance")

plt.xticks(p+0.2,x,rotation=10)
plt.show()
```
