```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,54,50,49,47]
min_t=[40,41,42,45,39,42,41]
avg_t = [45,46,47,44,43,45,48]


plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('weather')
plt.plot(days,max_t, label='max temp', marker="*")
plt.plot(days,min_t, label='min temp')
plt.plot(days,avg_t, label='avg temp')
plt.legend(shadow=True, fontsize='small')
plt.show()
```

```python
import numpy as np
company = ['google', 'facebook', 'amazon', 'tesla']
revenue=[15,20,30,70]
#xlabel
#ylabel
#title
xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.bar(xpos, revenue, label='revenue')
plt.legend()
plt.show()
```

```python
profit=[10,15,25,65]
plt.bar(xpos-0.2, revenue, width=0.4, label='revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='profit')
plt.xticks(xpos,company)
#XY label
#title
plt.legend(shadow=True)
plt.show()
```

```python
plt.barh(xpos, revenue)
plt.yticks(xpos,company)
plt.show()
```

```python
flights = sns.load_dataset('flights').pivot(index="month", columns='year')
flights
```

```python
plt.title("No. of passengers")
sns.heatmap(flights)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 15))
sns.heatmap(flights, fmt=".1f", annot=True, cmap="Blues", ax=ax)
ax.set_title("No. of passengers")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.rcParams.update({'font.size':10})
sns.heatmap(flights, fmt='d', annot=True, cmap='Blues', vmin=100, vmax=622, square=True,linewidth=1.4)
plt.show()
```

```python
plt.title("Flower")
sns.kdeplot(data = flw, x='sepal_length', y='sepal_width', shade=True, shade_lowest=False)
sns.lmplot(data = flw, x='sepal_length', y='sepal_width', fit_reg=False)
plt.show()
```

```python
plt.title("Flower")
sns.kdeplot(data = flw[flw.species=='setosa'], x='sepal_length', y='sepal_width', shade=True, shade_lowest=False, cmap='Reds')
sns.kdeplot(data = flw[flw.species=='virginica'], x='sepal_length', y='sepal_width', shade=True, shade_lowest=False, cmap='Blues')
sns.kdeplot(data = flw[flw.species=='versicolor'], x='sepal_length', y='sepal_width', shade=True, shade_lowest=False, cmap='Greens')
plt.show()
```
